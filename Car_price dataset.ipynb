```python
# Encoding dataset section in order to change the categorical value to numeric
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
# selecting the column to encode
car['Selling_type_n'] = le.fit_transform(car['Selling_type'])
car['Fuel_Type_n'] = le.fit_transform(car['Fuel_Type'])
car['Transmission_n'] = le.fit_transform(car['Transmission'])
```

```python
# call the data set
car
```

```python
# drop the colum categorical after the encoding   
car=car.drop(columns=['Fuel_Type','Selling_type','Transmission','Car_Name'])
```

```python
# call the dataset
car
```

```python
# checking for missing value
car.isnull().sum()
```

```python
# dropping the target variable and remains feature variable
X=car.drop(columns='Selling_Price')
X
```

```python
#selecting my target variable
y=car['Selling_Price']
y
```

```python
#importing library of training and test
from sklearn.model_selection import train_test_split
```

```python
## splitting  data into training and testing set section
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.2 ,random_state=42)
```

```python
#initialising the model and train in Randonregressor
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
```

```python
# initiallising the model
model.fit(X_train,y_train)
```

```python
# make prediction on the test 
y_predict=model.predict(X_test)
```

```python
y_predict
```

```python
#model evalutation  sections
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y_test,y_predict) 
```

```python
r2_score(y_test,y_predict)
```
